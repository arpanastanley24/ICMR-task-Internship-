# **QUESTION 2**

### Investigate if being Alcoholic is related to the marital status. Report the findings.

```python
# Create a cross-tabulation table
cross_tab = pd.crosstab(df['Alcohol'], df['Marital_status'])

# Print the cross-tabulation table
print(cross_tab)

# Plot a stacked bar plot
ax = sns.countplot(x='Marital_status', hue='Alcohol', data=df)

# Add count values to the plot
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')

#Labelling the plot
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.title('Relationship between Alcoholic and Marital Status')
plt.show()
```

# **QUESTION 3**

### Comment on the normality of “Income” variable using Kolmogorov Smirnov test and the QQ plot

```python
import scipy.stats as stats
import statsmodels.api as sm

# Extract the 'Income' variable
income = df['Income']

# Perform Kolmogorov-Smirnov test for normality
kstest_result = stats.kstest(income, 'norm')

# Print the test statistic and p-value
print("Kolmogorov-Smirnov test result:")
print("Test statistic:", kstest_result.statistic)
print("p-value:", kstest_result.pvalue)

alpha = 0.05
if kstest_result.pvalue>alpha:
  print('The income variable follows normal distribution')
elif kstest_result.pvalue<alpha:
  print('The income variable does not follow normal distribution')

# Create QQ plot
sm.qqplot(income, line='s')
plt.title('QQ Plot of Income')
plt.show()
```

# **QUESTION 4**

### For the given contingency form data investigate if the association between hours spent while revising (More than 8, Less than 8) is associated to the grade (Above average, Below average) or not?

                                | More than 8	 |   Less than 8
              __________________|_________________|_______________              
               Above Average	|     30    	  |       12
               Below Average	|     08	      |       29

### Report the findings.


```python
from scipy.stats import chi2_contingency

# Create a 2D array from the given data
observed = np.array([[30, 12],
                     [8, 29]])

# Perform chi-square test of independence
chi2, p, dof, expected = chi2_contingency(observed)

print("Chi-square statistic:", chi2)
print("P-value:", p)
print("Degrees of freedom:", dof)

alpha=0.05
if p>alpha:
  print('There is no significant association between hours spent and the grade')
elif p<alpha:
  print('There is significant association between hours spent and the grade')
```

### Using Crosssectional.xlsx data, compute appropriate summary measure for each of the following variables and represent the description of each of the variables graphically. Give reason for your choice

### (i)	Alcohol across Diabetes

```python
# Compute summary measures
alcohol_summary = df.groupby('Diabetes')['Alcohol'].describe()

# Print the summary measures
print(alcohol_summary)

# Rename values using rename() function in pandas
df['Diabetes'] = df['Diabetes'].replace({'0': 'No', '1': 'Yes'})
df['Alcohol'] = df['Alcohol'].replace({'0': 'No', '1': 'Yes'})

# Set the size of the plot
plt.figure(figsize=(5, 5.5))

# Plot a stacked bar plot
ax = sns.countplot(x='Alcohol', hue='Diabetes', data=df)

# Set x-axis labels
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])

# Get legend object
legend = ax.legend()

# Add title to legend
legend.set_title('Diabetes')

# Add count values to the plot
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')

#Labelling the plot
plt.xlabel('Alcohol')
plt.ylabel('Count')
plt.title('Alcohol across Diabetes')
plt.show()
```

### (ii)	Diet across gender

```python
# Compute summary measures
diet_summary = df.groupby(['Gender', 'Diet']).size().unstack()

# Print the summary measures
print(diet_summary)

# Rename values using rename() function in pandas
df['Gender'] = df['Gender'].replace({'1': 'Male', '2': 'Female'})
df['Diet'] = df['Diet'].replace({'1': 'Vegetarian', '2': 'Non-Vegetarian'})

# Set the size of the plot
plt.figure(figsize=(5.5, 5.5))

# Plot a stacked bar plot
ax = sns.countplot(x='Diet', hue='Gender', data=df)

# Set x-axis labels
plt.xticks(ticks=[0, 1], labels=['Vegetarian', 'Non-Vegetarian'])

# Get legend object
legend = ax.legend()

# Rename features in the legend
new_labels = ['Male', 'Female']
handles, _ = ax.get_legend_handles_labels()
ax.legend(handles, new_labels)

# Add count values to the plot
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')

#Labelling the plot
plt.xlabel('Diet')
plt.ylabel('Count')
plt.title('Diet across Gender')
plt.show()
```
